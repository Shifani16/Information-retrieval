```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KDTree
import joblib
import numpy as np
```

```python
data = pd.read_csv("cleaned_data.csv")
```

```python
tfidf = TfidfVectorizer(stop_words=None, lowercase=False, min_df=5)
train_tfidf = tfidf.fit_transform(data['abstract']).toarray()
```

```python
tfidf.vocabulary_
```

```python
data.isnull().sum()
```

```python
train_tfidf
```

```python
data['tfidf'] = list(train_tfidf)
data.head()
```

```python
kdtree = KDTree(train_tfidf)
```

```python
distance, idx = kdtree.query(data.tfidf[0].reshape(1, -1) ,k=3)

for i, value in enumerate(idx[0]):
    print(f'Judul: {data["judul"][value]}')
    print(f'URL : {data["link"][value]}')
    print(f'distance : {distance[0][i]}')
    print('\n')
```

```python
joblib.dump(tfidf, 'tfidf_model.pkl')

joblib.dump(kdtree, 'kdtree_model.pkl')
```
